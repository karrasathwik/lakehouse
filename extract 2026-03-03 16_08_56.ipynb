```python
class Extract:
    def __init__(self, spark):
        self.spark = spark
    def extract_Data(self):
        #intializing the catalag and schema
        silver_db = "lakehouse.sliver"
        bronze_db = "lakehouse.bronze"
        tables = self.spark.catalog.listTables(bronze_db)
        for i in tables:
            df = spark.read.table(f"{bronze_db}.{i.name}")
            #writing to sliver layer
            df.write.format("delta") \
                    .option("mode","append") \
                    .saveAsTable(f"{silver_db}.{i.name}")
            print(f"copied{i.name} to sliver layer")
extract_obj = Extract(spark)
extract_obj.extract_Data()
                    

```
