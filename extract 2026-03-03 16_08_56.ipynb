```python
%sql
CREATE CATALOG IF NOT EXISTS LakeHouse;
CREATE SCHEMA IF NOT EXISTS LakeHouse.Bronze;
CREATE SCHEMA IF NOT EXISTS LakeHouse.Sliver;
CREATE SCHEMA IF NOT EXISTS LakeHouse.Gold;
```

```python
#Reading the data from catalog
df = spark.read.table("lakehouse.bronze.sales_details")
df.show()
```

```python
bronze_db = "lakehouse.bronze"
sliver_db = "lakehouse.sliver"

tables = spark.catalog.listTables(bronze_db)

for i in tables:
    df = spark.read.table(f"{bronze_db}.{i.name}")
    df.write.format("delta").option("mode","append").saveAsTable(f"{sliver_db}.{i.name}")

                          
```
